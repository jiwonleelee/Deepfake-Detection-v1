```python
import pandas as pd

# 1. CSV 불러오기
df = pd.read_csv('/content/drive/MyDrive/HECTO/Dataset/CSV/total.csv')
```

```python
df
```

```python
# 2. 경로 치환 (구글 드라이브 경로 -> 로컬 content 경로)
# 기존 경로 예: /content/drive/MyDrive/HECTO/Dataset/Celeb_pp/frame_0.jpg
# 바뀔 경로 예: /content/data_local/Celeb_pp/frame_0.jpg
df['frame_path'] = df['frame_path'].str.replace('/content/drive/MyDrive/HECTO/Dataset', '/content/data_local/content/drive/MyDrive/HECTO/Dataset/Images')
```

```python
print(df['frame_path'])
```

```python
# 3. 수정된 데이터프레임을 새로운 CSV 파일로 저장
# index=False를 설정해야 불필요한 인덱스 번호가 파일에 포함되지 않습니다.
NEW_CSV_PATH = '/content/drive/MyDrive/HECTO/Dataset/CSV/total_local_path.csv'
df.to_csv(NEW_CSV_PATH, index=False)

print(f"✅ 경로 치환 완료 및 저장 성공: {NEW_CSV_PATH}")
```
